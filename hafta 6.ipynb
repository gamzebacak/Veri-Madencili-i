```python
jb=pd.read_csv("resources/jamesbond.csv", index_col="Film")
jb.sort_index(inplace=True)
```

```python
def degerlendir(row):
    aktor = row[1]
    butce = row[4]
    
    if(aktor=="Pierce Brosnan"):
        return "En İyi"
    elif(aktor=="Roger Moore" and butce > 40):
        return "Eğlenceli"
    else:
        return "Fikrim Yok"
```

```python
jb.apply(degerlendir,axis=1)
```
